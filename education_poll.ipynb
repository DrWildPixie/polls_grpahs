```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
plt.style.use('classic')
```

```python
df = pd.read_excel("poll2.xlsx")
```

```python
df
```

```python
df.columns
```

```python
df.percent=(df.voters /(df.voters.sum())*100).round(2)
df
```

```python
plt.bar(x=df['Education'], height=df.percent, color="black");
plt.title('Voters Education Percentage of r\Egypt Poll');
plt.ylabel('Percentage %');
plt.xlabel('Education Level');
plt.text(x=0, y=40,s='\nn=444\n', backgroundcolor='yellow');
plt.figure(figsize=(12,12));
```

```python
plt.barh(y=df['Education'], width=df.percent, color="black");
plt.title('Voters Education Percentage of r\Egypt Poll');
plt.xlabel('Percentage %');
plt.ylabel('Education Level');
plt.text(x=40,y=0,s='\nSample Size = 444\n', backgroundcolor='yellow')
plt.figure(figsize=(12,12));
```

```python
plt.figure(figsize=(12,12))
plt.pie(x=df.percent,labels=df.percent, colors=['white','purple','blue','green','yellow','black'], startangle=180)
plt.title('Voters Education Percentage of r\Egypt Poll');
plt.legend(df.Education)
plt.text(x=-1,y=1,s='\nSample Size = 444\n',backgroundcolor='yellow');
```
